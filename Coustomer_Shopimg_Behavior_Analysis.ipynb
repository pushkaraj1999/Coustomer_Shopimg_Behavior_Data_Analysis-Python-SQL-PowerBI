```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.columns
df= df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df.columns
```

```python
#Create a column age_group
labels=['Young-adult','Adult','Middel-age','Senior']
df['age_group']=pd.qcut(df['age'], q=4 , labels=labels)

df[['age','age_group']].head(10)
```

```python
# create new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Dropping promo code used column

df = df.drop('promo_code_used', axis=1)
```

```python
import urllib
from sqlalchemy import create_engine, text
import pandas as pd

# 1. Server setup (Using localhost shortcut for stability)
host = "DESKTOP-8IVIVJK\SQLEXPRESS"  

# 2. Step 1 Connection: Point to the default 'master' database
conn_str_master = (
    f"DRIVER={{ODBC Driver 17 for SQL Server}};"
    f"SERVER={host};"
    f"DATABASE=master;"  # Connect to system db first
    f"Trusted_Connection=yes;"
)
quoted_master = urllib.parse.quote_plus(conn_str_master)
engine_master = create_engine(f"mssql+pyodbc:///?odbc_connect={quoted_master}", isolation_level="AUTOCOMMIT")

# 3. Automatically create 'customer_behavior' if it doesn't exist
try:
    with engine_master.connect() as conn:
        # Check if database already exists
        check_db = conn.execute(text("SELECT db_id('customer_behavior')")).scalar()
        if check_db is None:
            conn.execute(text("CREATE DATABASE customer_behavior;"))
            print("Database 'customer_behavior' created successfully!")
        else:
            print("Database 'customer_behavior' already exists on your server.")
except Exception as e:
    print("Error checking/creating database:", e)

# 4. Step 2 Connection: Now point to your target database
conn_str_target = (
    f"DRIVER={{ODBC Driver 17 for SQL Server}};"
    f"SERVER={host};"
    f"DATABASE=customer_behavior;"  # Now this safely exists!
    f"Trusted_Connection=yes;"
)
quoted_target = urllib.parse.quote_plus(conn_str_target)
engine_target = create_engine(f"mssql+pyodbc:///?odbc_connect={quoted_target}")

# 5. Write your DataFrame to the target database
try:
    df.to_sql("customer", engine_target, if_exists="replace", index=False)
    print("DataFrame successfully written to 'customer' table!")
    
    # Verify by reading back 5 rows
    sample_df = pd.read_sql("SELECT TOP 5 * FROM customer;", engine_target)
    print("\nSample Data from SQL Server:")
    print(sample_df)
except Exception as e:
    print("Error writing data:", e)
```
